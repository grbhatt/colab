```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Data Import**

```python
df = pd.read_csv("/omni_min_2005may15-18.lst",sep = '\s+',header = None)
df.head(3)
```

```python
column_name = ['Year','Day','Hour','Minute','FMA','BX','BY','BZ',
               'Speed','ProtonD','Temp','Flow P','Electric F',
               'AE Index','AU Index','SYM/H']
df.columns = column_name
```

```python
df.tail(2)
```

```python
df['Time_period'] = pd.date_range(start = "2005-05-15", end = "2005-05-18 23:59",freq = 'T') 
```

```python
df.head(3)
```

```python
len(df.columns)
```

```python
fig,ax = plt.subplots(12, figsize  = (10,30))
j = 0
for i in column_name[4:16]:
  ax[j].plot(df['Time_period'],df[i])
  ax[j].set_ylabel(i)
  j+=1
ax[11].set_xlabel('Hour')
plt.show()
```
